```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("messydata.xlsx")
print(df.describe())
print(df.isna().sum())
df = df.drop(columns=["Unnamed: 0","Unnamed: 1","Sifra","Ucestalost","Ucestalost2","Deskripcija"])

df[['Datum', 'Vreme']] = df['Datum I vreme'].str.split(',', expand=True)
df = df.drop(columns=["Datum I vreme"])

df = df[df['Grad'] == 'BEOGRAD']

df['Datum'] = pd.to_datetime(df['Datum'], format='%d.%m.%Y')
df['Dan'] = df['Datum'].dt.day_name()
df['Mesec'] = df['Datum'].dt.month

broj_incidenata_po_opstini = df['Opstina'].value_counts()
df.to_parquet("cleaned_data2.parquet", index=False)
```
